**Business Questions Answered**

**1. For 'House of Cards', what is the show ID and who is the director of this show?**

```python
# Show all records of a particular item in any column
data[data['Title'].isin(['House of Cards'])]
```

```python
# (Alternate Method) show all records of a particular item in any column
data[data['Title'].str.contains('House of Cards')]
```

**2. In which year highest no. of TV shows & Movies were released? Show in bar Graph**

```python
# Since the Release date is in object data-type.
# create a new column (Date_N)

data['Date_N'] = pd.to_datetime(data['Release_Date'], errors='coerce')
```

```python
data.head()
```

```python
data.info()
```

```python
# It counts the occurrence of all individuals years in date column
data['Date_N'].dt.year.value_counts()
```

```python
# to show the bar graph
data['Date_N'].dt.year.value_counts().plot(kind='bar')
```

**3. How many Movies & TV shows are in the dataset? Show with bar graph**

```python
# group all unique items of a column and show their count
data.groupby('Category').Category.count()
```

```python
# in bar graph
sns.countplot(data['Category'])
```

**4. Show all the movies that were released in year 2020**

```python
# Create a new column that only contains Year
data['Year'] = data['Date_N'].dt.year
```

```python
data.head(2)
```

```python
# Filtering the Year column that only contains year = 2020
data[(data['Category'] == 'Movie') & (data['Year'] == 2020)]
```

**5. Show only title of all TV shows that were released in India only**

```python
# filtering the Category column that contains only TV shows, then filter country column that contains only India, then extract only title
data[(data['Category'] == 'TV Show') & (data['Country'] == 'India')] ['Title']
```

**7. Show all the records, where "Category is Movie and Type is Comedy" or "Country is United Kingdom"?**

```python
# Filtering
data[(data['Category'] == 'Movie') & (data['Type'] == 'Comedies') | (data['Country'] == 'United Kingdom')]
```

**8. In how many Movies/TV shows. Tom Cruise was cast?**

```python
# filtering
data[data['Cast'] == 'Tom Cruise']
# it shows no values having Tom Cruise Why?
# Answer: filtering doesnot apply because in Cast Column there are multiple items
```

```python
# Alternate Method using str.contains
data[data['Cast'].str.contains('Tom Cruise')]         # it shows an error there is null values
```

```python
# To correct the error
# Create a New Data Frame
data_new = data.dropna()      # it drops the rows that contains all or any missing values
```

```python
data_new.head(2)
```

```python
# now doing questions part
data_new[data_new['Cast'].str.contains('Tom Cruise')]
```
